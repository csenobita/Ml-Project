```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('data/data.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.drop(['street','statezip','yr_built','yr_renovated','waterfront','condition','view','sqft_basement'],axis=1,inplace=True)
```

```python
df.drop('date',axis=1,inplace=True)
```

```python
top_citys = df['city'].value_counts().nlargest(10).index
```

```python
df['city'] = df['city'].apply(lambda x : x if x in top_citys else "others" )
```

```python
df['city'].value_counts()
```

```python
X = df.drop(['price','country','city'],axis=1)
Y = df[['price']]
```

```python
# data prprocessing 
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python

from sklearn.svm import SVC
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python

int_coloumns=X.select_dtypes(['int','float']).columns
```

```python
int_coloumns
```

```python
# category_coloumns = X.select_dtypes('object').columns
```

```python
tr_x , tx_x , tr_y,tx_y =  train_test_split(X,Y,random_state=20,test_size=0.2)
```

```python
transformer_coloumns = ColumnTransformer(transformers=[
    #  ('cat',OneHotEncoder(),category_coloumns),
    ('num',StandardScaler(),int_coloumns),

]
)
```

```python
pipeline= Pipeline(
    steps=[
        ('preprocessor', transformer_coloumns),
        ('model',LinearRegression()),
          ]
)
```

```python
pipeline.fit(tr_x,tr_y)
```

```python
y=pipeline.predict(tx_x)
```
