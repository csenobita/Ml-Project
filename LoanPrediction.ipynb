# Import Library 

```python
import pandas as pd 
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

# CSV File Read

```python
df = pd.read_csv('data/loandataset.csv')
```

# Load Id Columns drop

```python
df.drop('Loan_ID',axis=1,inplace=True)
```

```python
# Data divided  X and Y
```

```python
X = df.drop('Loan_Status',axis=1)
Y = df['Loan_Status']
```

# Loan_Status Label Endcoding

```python
Y = Y.replace({'Y': 1, 'N': 0})
```

#  Data filling for most_frequent values 

```python
category_data= X.select_dtypes('object').columns
```

```python
cat_transfer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

#  Data filling for mean values using pipeline

```python
numbers_data=  X.select_dtypes(['int','float']).columns
```

```python
num_transfer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy="mean")),
    ('scaler',StandardScaler())
])
```

# Pipe Line setup for missing values 

```python
preprocessor =  ColumnTransformer(transformers=[
    ('num',num_transfer,numbers_data),
    ('cat',cat_transfer,category_data)
])
```

# Model setup

```python
pipeline= Pipeline(steps=[
    ('preprocessing',preprocessor),
    ('model',LogisticRegression())
]
)
```

# Model train and text data

```python
x_train,x_text,y_train,y_text = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

# Model Traing 

```python
pipeline.fit(x_train,y_train)
```

# Model Texting

```python
y_pred= pipeline.predict(x_text)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

# Model Accouracy

```python
accuracy_score(y_pred,y_text)
```

```python
r2_score(y_pred,y_text)
```
