# Library Imorting 

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix,ConfusionMatrixDisplay
```

# Raw Data import

```python
with open("Parkinsons disease raw data.txt",'r') as f:
    lines = f.readlines()
```

```python
text = [line.strip() for line in lines if line.strip() != "" ]
```

# Row and columns created

```python
div = 1
data = []
columns = [0]
for i in range(1,4680):
    if i / div == 24 : 
        div = div +1
        columns.append(i)

for i in columns : 
    temp=[text[i]]
    for j in range(1,24):
        temp.append(text[i+j])
    data.append(temp)
```

# Data Frame Created

```python
df  = pd.DataFrame(data)
```

# Data Corr chacking

```python
corr = df.drop(0,axis=1)
```

```python
corr = corr.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True)
plt.show()
```

# Inputs and target Data create 

```python
x = df.drop([0,17],axis=1)
y = df[17]
```

# Train and Test Data split

```python
train_x , test_x , train_y, test_y = train_test_split(x,y,stratify=y,random_state=42,test_size=0.2)
```

# Logistic Regression Model create and train 

```python
model = LogisticRegressionCV(max_iter=100000)
model.fit(train_x,train_y)
pred = model.predict(test_x)
print('Model Acccuracy : ' , accuracy_score(pred,test_y))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(pred,test_y)).plot()
plt.show()
```

# RandomForestClassifier Model create and train 

```python
model = RandomForestClassifier(random_state=20,n_estimators=1000)
model.fit(train_x,train_y)
pred = model.predict(test_x)
print('Model Acccuracy : ' , accuracy_score(pred,test_y))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(pred,test_y)).plot()
plt.show()
```

# Decision Tree Classifier Model create and train 

```python
model = DecisionTreeClassifier(criterion='log_loss')
model.fit(train_x,train_y)
pred = model.predict(test_x)
print('Model Acccuracy : ' , accuracy_score(pred,test_y))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(pred,test_y)).plot()
plt.show()
```

# SVC Model create and train 

```python
model = SVC(max_iter=10000,random_state=20)
model.fit(train_x,train_y)
pred = model.predict(test_x)
print('Model Acccuracy : ' , accuracy_score(pred,test_y))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(pred,test_y)).plot()
plt.show()
```
